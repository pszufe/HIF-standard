```python
import xgi
```

## Undirected Hypergraph

```python
H = xgi.load_xgi_data("email-enron")
xgi.write_hif(H, "data/email-enron.json")
```

```python
H_hif = xgi.read_hif("data/email-enron.json")
print(H_hif.nodes)
H_hif = xgi.read_hif("data/email-enron.json", nodetype=int)
print(H_hif.nodes)
H_hif.cleanup()
```

```python
pos = xgi.pairwise_spring_layout(H_hif)
pos = xgi.pca_transform(pos)
xgi.draw(H_hif, pos=pos, max_order=3)
```

```python
# get all the email addresses for every node
H_hif.nodes.attrs("name").asdict()
```
