```python
import hypergraphx as hgx
from hypergraphx.viz.draw_hypergraph import draw_hypergraph as hgx_draw
import HAT
import hypernetx as hnx
import numpy as np
import networkx as nx
import nx_hif
import xgi
```

```python
filename = "data/hif-issue-27.json"
H_hgx = hgx.readwrite.load_hif(filename)
hgx_draw(H_hgx)
print(H_hgx.get_nodes())
print(H_hgx.get_edges())
```

```python
H_hnx = hnx.from_hif(filename=filename)
hnx.draw(H_hnx)
print(list(H_hnx.nodes))
print(list(H_hnx.edges))
```

```python
G_nx_hif = nx_hif.read_hif(filename)
nx.draw_networkx(G_nx_hif)
print(G_nx_hif.nodes)
print(G_nx_hif.edges)
```

```python
H_xgi = xgi.read_hif(filename)
xgi.draw(H_xgi)
print(H_xgi.nodes)
print(H_xgi.edges)
```
