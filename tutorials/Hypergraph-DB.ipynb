# Using HIF datasets with Hypergraph-DB

[Hypergraph-DB](https://github.com/hypergraph-db/hypergraph-db) is a Lightweight, Fast, and Modern Hypergraph Database with Hypergraph Visualization.

Hypergraph-DB is a lightweight, flexible, and high-performance Python library for modeling and managing hypergraphs ‚Äî advanced graph structures where edges (hyperedges) can connect any number of vertices. Unlike traditional graphs that only support pairwise relationships, hypergraphs naturally represent complex multi-way relationships found in real-world scenarios.

This notebook demonstrates:
1. Loading a hypergraph from a HIF format JSON file using `load_from_hif(filepath)`
2. Saving a hypergraph to a HIF format JSON file using `save_as_hif(filepath)`

## Load HIF File

The `load_from_hif(filepath)` method loads a hypergraph from a HIF format JSON file.

```python
from hyperdb import HypergraphDB
```

```python
# Load a different HIF file - publications dataset
hg = HypergraphDB()
hg.load_from_hif("data/publications_main_component.hif.json")

print(hg.num_v)
print(hg.num_e)
print(hg.all_v)
```

## üé® Interactive Visualization 

Explore your hypergraphs visually with the built-in web-based visualization tool:


```python
hg.draw()
```

## Save HIF File

The `save_as_hif(filepath)` method saves a hypergraph to a HIF format JSON file.

```python
# Create a simple hypergraph example
hg_example = HypergraphDB()

# Add nodes with attributes
hg_example.add_v(1, {"name": "Alice", "institution": "University A"})
hg_example.add_v(2, {"name": "Bob", "institution": "University B"})
hg_example.add_v(3, {"name": "Charlie", "institution": "University A"})

# Add edges with attributes
hg_example.add_e((1, 2), {"relation": "coauthor", "year": 2023})
hg_example.add_e((2, 3), {"relation": "coauthor", "year": 2024})
hg_example.add_e((1, 2, 3), {"relation": "team", "year": 2023})
```

```python
# Save the hypergraph to a HIF format JSON file
hg_example.save_as_hif("data/example_hypergraph.hif.json")
print("Hypergraph saved successfully!")
```
