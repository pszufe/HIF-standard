## Demo: Generating a Hypergraph with Semantic Labels and Exporting to HIF

In this section, we manually create a hypergraph with specific node attributes (e.g., `author_name`) and edge attributes (e.g., `paper_title`). We then demonstrate how to use the `node_label` and `edge_label` parameters to export this graph into a semantic, human-readable HIF JSON file, where nodes and edges are indexed by meaningful names instead of integer IDs.

```python
# ==========================================
# Step 1: Create a Hypergraph
# ==========================================
print("--- 1. Creating a Hypergraph with Attributes ---")


v_props = [
    {"author_name": "Alice", "institution": "Univ A"},   # Node 0
    {"author_name": "Bob", "institution": "Univ B"},     # Node 1
    {"author_name": "Charlie", "institution": "Univ A"}  # Node 2
]

e_list = [
    (0, 1), 
    (1, 2)
]

e_props = [
    {"paper_title": "Graph Theory 101", "year": 2023},    # Edge 0
    {"paper_title": "HIF Format Guide", "year": 2024}     # Edge 1
]

H = eg.Hypergraph(
    num_v=3,
    e_list=e_list,
    v_property=v_props,
    e_property=e_props
)

print(f"Hypergraph created with {H.num_v} nodes and {len(H.e[0])} edges.")
print(f"Node 0 Props: {H.v_property[0]}")
print(f"Edge 0 Props: {H.e_property['main'][0]}")


# ==========================================
# Step 2: Export
# ==========================================
print("\n--- 2. Exporting with Semantic Labels ---")
output_file = "demo_semantic_final.hif.json"

eg.hypergraph_to_hif(
    H, 
    output_file, 
    node_label="author_name",  
    edge_label="paper_title"  
)
print(f"Exported to {output_file} successfully.")


# ==========================================
# Step 3: Verify JSON
# ==========================================
print("\n--- 3. Inspecting the Generated JSON ---")
with open(output_file, 'r') as f:
    data = json.load(f)

# The verification node ID has changed to "Alice"
print("First Node in JSON:", json.dumps(data['nodes'][0], indent=2, ensure_ascii=False))
# The verification edge ID has changed to "Graph Theory 101" 
print("First Edge in JSON:", json.dumps(data['edges'][0], indent=2, ensure_ascii=False))


# ==========================================
# Step 4: Import and Index
# ==========================================
print("\n--- 4. Importing and Indexing by Name ---")

H_new = eg.hif_to_hypergraph(
    filename=output_file,
    node_label="author_name", 
    edge_label="paper_title"
)

if hasattr(H_new, 'node_label_index'):
    print("Using built-in label index...")
    name_map = H_new.node_label_index
else:
    name_map = {}
    for i in range(H_new.num_v):
        name = H_new.v_property[i].get('name') 
        if name:
            name_map[name] = i

target_author = "Alice"

if target_author in name_map:
    nid = name_map[target_author]
    props = H_new.v_property[nid]
    print(f"Found Author: '{target_author}'")
    print(f"-> Internal ID: {nid}")
    print(f"-> Institution: {props.get('institution')}")
else:
    print(f"Author '{target_author}' not found.")
```
